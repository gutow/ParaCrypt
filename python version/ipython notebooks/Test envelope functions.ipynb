```python
from encrypcheckutil import *
import matplotlib.pyplot as plt
```

```python
yvalues = np.empty(2048)
xvalues = np.empty(2048)
for i in range (0,2048):
    xvalues[i] = 200*i/2048
    yvalues[i]= np.sin(xvalues[i])+np.sin(xvalues[i]*1.1)
plt.plot(xvalues,yvalues)
plt.show()
```

```python
upperenvx, upperenvy, lowerenvx, lowerenvy=envelope(xvalues,yvalues)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(xvalues,yvalues,'.',label='data')
plt.plot(upperenvx,upperenvy,'-',label='upper envelope')
plt.plot(lowerenvx,lowerenvy,'-',label='lower envelope')
plt.legend()
plt.show()
```
